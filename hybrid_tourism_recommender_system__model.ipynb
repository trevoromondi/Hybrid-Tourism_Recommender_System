```python
from scipy.sparse import data
#preprocessing data
import string 
import re 

data_cb_cleaned['title_review'] = data_cb_cleaned['title_review'].astype(str)
data_cb_cleaned['city'] = data_cb_cleaned['city'].astype(str)


#removing underscores
#data_cb_cleaned = data_cb_cleaned.replace("_" , "  ") 

#2. feature merge
data_cb_cleaned['metadata'] = data_cb_cleaned['title_review'].str.cat(data_cb_cleaned['city'], sep=', ').to_frame()

#3. lower case
def lower_case(text):
  return text.casefold()

data_cb_cleaned['metadata'] = data_cb_cleaned['metadata'].apply(lower_case)

#4. remove number
def remove_number(text):
    return  re.sub(r"\d+", "", text)

data_cb_cleaned['metadata'] = data_cb_cleaned['metadata'].apply(remove_number)

#5. remove punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans("","",string.punctuation))

data_cb_cleaned['metadata'] = data_cb_cleaned['metadata'].apply(remove_punctuation)

data_cb_cleaned.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_distances

!pip install nltk
import nltk
nltk.download('punkt')

class rec_cb:
  def __init__(self, data, content_col):
    self.df = data_cb_cleaned
    self.content_col = content_col
    self.encoder = None
    self.bank = None

  def fit(self):
    self.encoder = TfidfVectorizer(stop_words='english', tokenizer=word_tokenize)
    self.bank = self.encoder.fit_transform(self.df[self.content_col])
  
  def recommend(self, idx, topk=10):
    content = self.df.loc[idx, self.content_col]
    code = self.encoder.transform([content])
    dist = cosine_distances(code, self.bank)
    rec_idx = dist.argsort() [0, 0:(topk+1)]
    recom_cb = self.df.loc[rec_idx].restaurant_name
    df_cosine = pd.DataFrame({'restaurant_name': recom_cb, 'idx' : rec_idx})
    df_cosine['cosine_distance'] = dist[0,df_cosine.idx]
    self.df['cosine_distance'] = df_cosine['cosine_distance']
    return self.df.loc[self.df.cosine_distance <= 0.6]
```

```python
recomendasi_cb = rec_cb('data_cb_cleaned', content_col = 'metadata')
recomendasi_cb.fit()
```

```python
#Preprocessing

#1. Pilih Fitur yang dibutuhkan
df_cf = pd.DataFrame({
    'review_ids' : df_cf_pred_cleaned["review_ids"],
    'restaurant_name' : df_cf_pred_cleaned["restaurant_name"],
    'rating_review' : df_cf_pred_cleaned["rating_review"]
})
df_cf.head()
```

```python
#Preprocessing

#2. check missing value
df_cf.isna().sum()
```

```python
!pip install scikit-surprise
from surprise import SVD, Dataset, Reader
from surprise.model_selection import cross_validate

class rec_cf:
  def __init__(self, data):
    self.df = df_cf
    self.all_destinations = self.df.restaurant_name.unique()
    self.model = None
  
  def fit(self):
    data = Dataset.load_from_df(df_cf, Reader())
    trainset = data.build_full_trainset()
    self.model = SVD()
    self.model.fit(trainset)

  def recommend(self, review_ids, topk = 77):
    rated = self.df[self.df.review_ids == review_ids].restaurant_name
    not_rated = [restaurant_name for restaurant_name in self.all_destinations if restaurant_name not in rated]
    score = [self.model.predict(review_ids, restaurant_name).est for restaurant_name in not_rated]
    result = pd.DataFrame({'restaurant_name': not_rated, 'prediksi_rating_cf' : score})
    return result.head(topk)
```

```python
rekomendasi_cf = rec_cf('data_cf')
rekomendasi_cf.fit()
```

```python
df_prediksi2 = rekomendasi_cf.recommend(review_ids = input_a)
```
